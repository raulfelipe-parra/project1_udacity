# Rio de Janeiro Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_calendar = pd.read_csv('./dataset/calendar.csv')
df_listing = pd.read_csv('./dataset/listings.csv')
```

```python
df_calendar.head()
```

```python
df_listing.head()
```

## Question 1

### Which are the top commun host neigbourhood?

```python
host_neighbourhood_count = df_listing.host_neighbourhood.value_counts()

print(host_neighbourhood_count.head(15))
```

```python
(host_neighbourhood_count.head(15) / df_listing.shape[0]).plot(kind = 'bar');
plt.title('Top 15 Most Common Host Neighbourhood');
plt.grid(True);
```

## Question 2

### What is the impact of Brazil holidays in terms of price?

```python
df_listing[df_listing['id'] == 1174007]
```

```python
df_calendar[df_calendar['listing_id'] == 1174007]
```

```python
list(df_listing.columns)
```

```python
df_calendar['room_type'] = df_calendar.listing_id.map(df_listing.set_index('id')['room_type'].to_dict())
```

```python
df_calendar[df_calendar['listing_id'] == 1174007]
```

```python
df_calendar.info()
```

```python
df_calendar['date'] = pd.to_datetime(df_calendar['date'])
```

```python
df_calendar['price'] = df_calendar['price'].str.replace('$', '')

df_calendar['price'] = df_calendar['price'].str.replace(',', '')

df_calendar['price'] = df_calendar['price'].astype(float)
```

```python
#List of type of room
list(df_calendar['room_type'].unique())
```

```python
df_calendar_privateRoom = df_calendar[df_calendar['room_type'] == 'Private room']

df_calendar_entireHome = df_calendar[df_calendar['room_type'] == 'Entire home/apt']

df_calendar_sharedRoom = df_calendar[df_calendar['room_type'] == 'Shared room']

df_calendar_hotelRoom = df_calendar[df_calendar['room_type'] == 'Hotel room']
```

```python
df_calendar_privateRoom.groupby('date').price.mean().plot()
plt.title('Private Room Price Series')
plt.grid(True)
```

```python
df_calendar_entireHome.groupby('date').price.mean().plot()
plt.title('Entire Home Price Series')
plt.grid(True)
```

```python
df_calendar_sharedRoom.groupby('date').price.mean().plot()
plt.title('Private Room Price Series')
plt.grid(True)
```

```python
df_calendar_hotelRoom.groupby('date').price.mean().plot()
plt.title('Private Room Price Series')
plt.grid(True)
```
